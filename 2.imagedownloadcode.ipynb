```python
import json
import os
import requests

API_KEY = 'c047480430ce46ceadcf452e7472236e'
language = 'hi'
pageSize = 100
totalNewsNeeded = 1000
fetchedArticlesFilePath = "/content/fetched_articles.json"#os.path.join(os.path.dirname(__file__), 'fetched_articles.json')

def save_news_to_file(news):
    fileName = 'news.txt'
    filePath = os.path.join(os.path.dirname("/content/"), fileName)

    fetchedArticles = set()
    if os.path.exists(fetchedArticlesFilePath):
        with open(fetchedArticlesFilePath, 'r') as file:
            fetchedArticles = set(json.load(file))

    news = [article for article in news if article['headline'] not in fetchedArticles]

    with open(filePath, 'a', encoding='utf-8') as file:
        for article in news:
            file.write(f"{article['headline']}\n{article['imageUrl']}\n\n")
            fetchedArticles.add(article['headline'])

    with open(fetchedArticlesFilePath, 'w', encoding='utf-8') as file:
        json.dump(list(fetchedArticles), file, ensure_ascii=False, indent=4)

def fetch_news(keyword):
    try:
        url = 'https://newsapi.org/v2/everything'
        params = {
            'q': keyword,
            'language': language,
            'apiKey': API_KEY,
            'pageSize': pageSize
        }
        response = requests.get(url, params=params)
        response.raise_for_status()
        data = response.json()

        news = [{'headline': article['title'], 'imageUrl': article['urlToImage']} for article in data['articles']]

        if len(news) == 0:
            raise Exception('No news articles found.')

        save_news_to_file(news)
    except Exception as e:
        print('Error fetching or saving news:', e)

keyword = 'कोई'  # Example keyword, replace with your desired keyword
fetch_news(keyword)
```

```python

import xlsxwriter

def create_excel_from_text(text_file, excel_file):
    # Create a new Excel workbook and add a worksheet
    workbook = xlsxwriter.Workbook(excel_file)
    worksheet = workbook.add_worksheet()

    # Add headers to the columns
    worksheet.write('A1', 'Text')
    worksheet.write('B1', 'Image Link')

    # Open the text file and read lines
    with open(text_file, 'r', encoding='utf-8') as file:
        lines = file.readlines()

    # Iterate through the lines of the text file
    row = 1  # Start from the second row to skip headers
    text = ''
    for line in lines:
        line = line.strip()  # Remove leading/trailing whitespaces
        if line:  # Non-empty line
            if not text:  # If text is empty, it's a new text
                text = line
            else:  # If text is not empty, it's an image link
                image_link = line
                # Write data to the Excel worksheet
                worksheet.write(row, 0, text)
                worksheet.write(row, 1, image_link)
                row += 1  # Move to the next row
                text = ''  # Reset text for the next iteration

    workbook.close()
    print(f"Excel file '{excel_file}' has been created successfully.")

# Example usage
text_file = '/content/su.txt'
excel_file = 'output.xlsx'
create_excel_from_text(text_file, excel_file)
```

```python
import requests
from openpyxl import load_workbook
import os

# Create a folder to store downloaded images
if not os.path.exists('new'):
    os.makedirs('new')

# Load the Excel sheet with only image links
wb = load_workbook('/content/output.xlsx')
sheet = wb.active

# Start numbering images from 50001
image_id = 1

# Iterate through each row in the sheet and download the image
for row in sheet.iter_rows(values_only=True):
    image_url = row[1]  # Assuming the image links are in the first column

    # Check if the cell contains a valid URL
    if image_url and image_url.startswith('http'):
        try:
            # Download the image
            response = requests.get(image_url)
            if response.status_code == 200:
                # Save the image with the assigned ID
                with open(f'new/{image_id}.jpg', 'wb') as f:
                    f.write(response.content)
                print(f'new{image_id}')
                image_id += 1
            else:
                print(f'Failed to download image {image_url}')
        except Exception as e:
            print(f'Error downloading image {image_url}: {e}')

# Save the Excel file (optional)
# wb.save('image_links.xlsx')
```
